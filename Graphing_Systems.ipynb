```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10

#define how many points to plot
points = 10*(xmax-xmin)

#define the array of x value once
x = np.linspace(xmin,xmax,points)

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

#line 1
y1 = x+6.5
plt.plot(x , y1)
plt.fill_between(x , y1, ymax, facecolor='darkblue')
#line 2
y2 = x+4.5
plt.plot(x, y2,'-')
plt.fill_between(x, y2, y1, facecolor ='white')

#line 3
y3 = x-2.5
plt.plot(x, y3)
plt.fill_between(x, y3, y2, facecolor='darkred')

#line 4
y4 = x-4.5
plt.plot(x, y4,'-')
plt.fill_between(x, y4, y3, facecolor='white')

#line 5
y5 = x-20
plt.plot(x, y5)
plt.fill_between(x, y5, y4, facecolor='darkgreen')
plt.title('FOR HAGE-LO LO')

plt.show()
```

# **ineractive** graph

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

#Define the graphine function
def f(m, b, zoom):
    xmin = -zoom
    xmax = zoom
    ymin = -zoom
    ymax = zoom
    #define how many points to plot
    points = 10*(xmax-xmin)
    #define the array of x value once
    x = np.linspace(xmin,xmax,points)
    fig, ax = plt.subplots()
    plt.axis([xmin,xmax,ymin,ymax]) # window size
    plt.plot([xmin,xmax],[0,0],'black') # black x axis
    plt.plot([0,0],[ymin,ymax], 'black') # black y axis
    plt.title('y = mx + 1')

    plt.plot(x, m*x+b)
    plt.show()

#set up the sliders
interactive_plot = interactive(f, m=(-9, 9), b=(-9,9), zoom=(1,100))
interactive_plot
```

```python
#import Meteostat libraby and dependencies
from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Point, Daily

# Documentation
#https://dev.meteostat.net/python/daily.html#data-structure

#set time period
start = datetime(2022, 1, 1)
end = datetime(2022, 7, 31)

#creape point for vancouver or philadelphia, PA
#vancouver = Point(49.2497, -123.1193, 70)
philly = Point(39.97,-75.13,25)

#get daily for 2018
data = Daily(philly, start, end)
data = data.fetch()

#plot line chart including average , minimum and maximum temperature
data.plot(y=['tavg', 'prcp','snow'])
plt.show
```
